```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import IBMQ
from qiskit_experiments.database_service import DbExperimentDataV1 as DbExperimentData
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q-research', group='tokyo-1', project='main')
backend = provider.get_backend('ibmq_quito')
```

```python
li_exp_data = DbExperimentData.load("1d33e126-cc14-4908-ade2-e21a5881d754", provider.service("experiment"))
pe_exp_data = DbExperimentData.load("1ad15d3f-58ea-4d94-ac49-a1c860c3f2fe", provider.service("experiment"))
cx_exp_data = DbExperimentData.load("93a1e2b4-c80d-49d5-adac-13130a48abfb", provider.service("experiment"))
```

```python
phi_values = li_exp_data.metadata['phi_values']
plt.scatter(phi_values, li_exp_data.analysis_results()[0].value, label='Linearized')
plt.scatter(phi_values, pe_exp_data.analysis_results()[0].value, label='Pulse-efficient')
plt.scatter(phi_values, cx_exp_data.analysis_results()[0].value, label='Double CX')
plt.xlabel(r'$\phi$', fontsize='x-large')
plt.ylabel(r'$R_{ZZ}(\phi)$ process fidelity', fontsize='x-large')
plt.legend()
```
